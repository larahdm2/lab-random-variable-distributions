```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Use the numerical variables from our dataset this week.

```python
numerical = pd.read_csv('numerical_230823.csv')
```

# 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
for col in numerical.columns:
    sns.distplot(numerical[col])
    plt.show()
```

None of the columns fit a normal distribution. Considering https://www.w3schools.com/python/numpy/numpy_random_exponential.asp, monthly_premium_auto and total_claim_amount could fit an exponential distribution.

# 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
def log_transform_clean(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return 0
```

```python
for col in numerical.columns:
    
    numerical[col] = list(map(log_transform_clean, numerical[col]))
    numerical[col] = numerical[col].fillna(np.mean(numerical[col]))
    
    sns.distplot(numerical[col])
    plt.show()
```

The result has changed but still none of the plot fit a normal distribution.
